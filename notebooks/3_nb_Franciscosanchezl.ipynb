```python
def cambio_mes (param):
    if param == "Ene":
        param = "Jan"
    if param == "Feb":
        param = "Feb"
    if param == "Mar":
        param = "Mar"
    if param == "Abr":
        param = "Apr"
    if param == "May":
        param = "May"
    if param == "Jun":
        param = "Jun"
    if param == "Jul":
        param = "Jul"
    if param == "Ago":
        param = "Aug"
    if param == "Sep":
        param = "Sep"
    if param == "Oct":
        param = "Oct"
    if param == "Nov":
        param = "Nov"
    if param == "Dic":
        param = "Dec"
        
    return param
```

```python
df_sii['Mes_en'] = df_sii["Mes"].apply(cambio_mes)
df_sii
```

```python
df_sii['Fecha'] =  (df_sii['Día'].apply(str) +"-"+ df_sii['Mes_en'] +"-"+  df_sii['year'].apply(str))
df_sii
```

```python
from datetime import datetime

df_sii['Fecha'] = pd.to_datetime(df_sii['Fecha'], format='%d-%b-%Y')
df_sii
```

```python
df_sii['dia_de_la_semana'] = df_sii['Fecha'].dt.day_name() 
df_sii
```

```python
def dia_semana (dia):
    if dia == "Monday":
        dia = "Lunes"
    if dia == "Tuesday":
        dia = "Martes"
    if dia == "Wednesday":
        dia = "Miércoles"
    if dia == "Thursday":
        dia = "Jueves"
    if dia == "Friday":
        dia = "Viernes"
    if dia == "Saturday":
        dia = "Sábado"
    if dia == "Sunday":
        dia = "Domingo"
        
    return dia
```

```python
df_sii['dia_de_la_semana'] = df_sii["dia_de_la_semana"].apply(dia_semana)
df_sii
```

```python
# Fechas de cumpleanios
fecha1_mes = ['04']
fecha1_dia = ['18']

fecha2_mes = ['06']
fecha2_dia = ['23']

fecha3_mes = ['03']
fecha3_dia = ['22']

fecha4_mes = ['01']
fecha4_dia = ['10']

fecha5_mes = ['01']
fecha5_dia = ['23']

df_cumpleanios = df_bc_all[ (df_bc_all.mes.isin(fecha1_mes) & df_bc_all.dia.isin(fecha1_dia)) |
                            (df_bc_all.mes.isin(fecha2_mes) & df_bc_all.dia.isin(fecha2_dia)) |
                            (df_bc_all.mes.isin(fecha3_mes) & df_bc_all.dia.isin(fecha3_dia)) |
                            (df_bc_all.mes.isin(fecha4_mes) & df_bc_all.dia.isin(fecha4_dia)) |
                            (df_bc_all.mes.isin(fecha5_mes) & df_bc_all.dia.isin(fecha5_dia))
                          ]

df_cumpleanios = df_cumpleanios[(df_cumpleanios['anio'] != '1982') & (df_cumpleanios['anio'] != '2022')]
df_cumpleanios
```

```python
df_cumpleanios_grupo = df_cumpleanios.groupby(by=["anio"]  ,as_index = False)["1.Dólar observado"].mean()
df_cumpleanios_grupo
```
